## Subset:
Problem Description

Given a set of distinct integers, S, return all possible subsets. 

### NOTE:
Elements in a subset must be in non-descending order.

The solution set must not contain duplicate subsets.

Also, the subsets should be sorted in ascending ( lexicographic ) order.

The list is not necessarily sorted.
 
### Problem Constraints
1 <= |S| <= 16

INTMIN <= S[i] <= INTMAX

### Input Format
First and only argument of input contains a single integer array S.

### Output Format
Return a vector of vectors denoting the answer.

### Example Input
 S = [1, 2, 3]

### Example Output
[
 
 []
 
 [1]
 
 [1, 2]
 
 [1, 2, 3]
 
 [1, 3]
 
 [2]
 
 [2, 3]
 
 [3]

]

### Example Explanation
 You can see that these are all possible subsets.

```python
class Solution:
    # @param A : list of integers
    # @return a list of list of integers
    def subsets(self, a):
        a.sort()
        ans = self.subsetsHelper(a, 0, [])
        ans.sort()
        return ans
    def subsetsHelper(self, a, i, ans):
        if i == len(a):
            ans.append([])
            return ans        
        so = self.subsetsHelper(a, i+1, ans)
        for x in range(len(so)):
            ans.append([a[i]] + so[x])
        return ans
o = Solution()
print(o.subsets([1,2,3]))
```

```python
class Solution:
    # @param A : list of integers
    # @return a list of list of integers
    def subsets(self, s):
        s.sort()
        r = [[]]
        curr = []
        for e in s:
            r += [x+[e] for x in r]
        return sorted(r)
o = Solution()
print(o.subsets([-98, 54, -52, 15, 23, -97, 12, -64, 52, 85]))
```

```python
class Solution:
    def subsets(self, A):
        ans = self.subsets_helper(A, 0, [], [])
        if not ans:
            return []
        ans.sort()
        return ans
    def subsets_helper(self, array, index, current_subset, ans_array):
        if index == len(array):
            ans_array.append(current_subset)
            return
        self.subsets_helper(array, index + 1, current_subset, ans_array)
        self.subsets_helper(array, index + 1, current_subset + [array[index]], ans_array)
        return ans_array
    
o = Solution()
print(o.subsets([ ]))
```
