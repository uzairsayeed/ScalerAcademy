### Dungeon Princess:
Problem Description

The demons had captured the princess and imprisoned her in the bottom-right corner of a dungeon. The dungeon consists of M x N rooms laid out in a 2D grid. Our valiant knight was initially positioned in the top-left room and must fight his way through the dungeon to rescue the princess.

The knight has an initial health point represented by a positive integer. If at any point his health point drops to 0 or below, he dies immediately.

Some of the rooms are guarded by demons, so the knight loses health (negative integers) upon entering these rooms; other rooms are either empty (0's) or contain magic orbs that increase the knight's health (positive integers).

In order to reach the princess as quickly as possible, the knight decides to move only rightward or downward in each step.

Given a 2D array of integers A of size M x N. Find and return the knight's minimum initial health so that he is able to rescue the princess.
### Problem Constraints
1 <= M, N <= 500

-100 <= A[i] <= 100
### Input Format
First and only argument is a 2D integer array A denoting the grid of size M x N.
### Output Format
Return an integer denoting the knight's minimum initial health so that he is able to rescue the princess.
## Example Input
### Input 1:
 A = [ 
       
       [-2, -3, 3],
       
       [-5, -10, 1],
       
       [10, 30, -5]
     
     ]
### Input 2:
 A = [ 
       
       [1, -1, 0],
       
       [-1, 1, -1],
       
       [1, 0, -1]
     
     ]
## Example Output
### Output 1:
 7
### Output 2:
 1
## Example Explanation
### Explanation 1:
 Initially knight is at A[0][0].
 
 If he takes the path RIGHT -> RIGHT -> DOWN -> DOWN, the minimum health required will be 7.
 
 At (0,0) he looses 2 health, so health becomes 5.
 
 At (0,1) he looses 3 health, so health becomes 2.
 
 At (0,2) he gains 3 health, so health becomes 5.
 
 At (1,2) he gains 1 health, so health becomes 6.
 
 At (2,2) he looses 5 health, so health becomes 1.
 
 At any point, the health point doesn't drop to 0 or below. So he can rescue the princess with minimum health 7.
### Explanation 2:
 Take the path DOWN -> DOWN ->RIGHT -> RIGHT, the minimum health required will be 1.

```python
# RecursiveApproach 
class Solution:
    # @param A : list of list of integers
    # @return an integer
    def calculateMinimumHP(self, a):
        r, c = len(a), len(a[0])
        if a[r-1][c-1] >=0:
            final_health = 1
        else:
            final_health = 1 - a[r-1][c-1]
        return self.helperFn(a, r-1, c-1, final_health)
    
    def helperFn(self, a, i, j, h):
        if i == 0 and j == 0:
            return h
        if i == 0:
            x = a[i][j-1]
            if x >= h:
                h_x = 1
            else:
                h_x = h - x
            return self.helperFn(a, i, j-1, h_x)
        elif j == 0:
            y = a[i-1] [j]
            if y >= h:
                h_y = 1
            else:
                h_y = h - y
            return self.helperFn(a, i-1, j, h_y)        
        x, y = a[i-1][j], a[i][j-1]
        if x >= h:
            h_x = 1
        else:
            h_x = h - x
        if y >= h:
            h_y = 1
        else:
            h_y = h - y
        return min(self.helperFn(a, i-1, j, h_x), self.helperFn(a, i, j-1, h_y))
        
o = Solution()
A = [

   [-2, -3, 3],

   [-5, -10, 1],

   [10, 30, -5]

 ]
print(o.calculateMinimumHP(A))
```

```python
# RecursiveApproach + Memoization --> TopDown DP 
class Solution:
    # @param A : list of list of integers
    # @return an integer
    def calculateMinimumHP(self, a):
        r, c = len(a), len(a[0])
        dp = [[0 for j in range(c)]for i in range(r)]
        if a[r-1][c-1] >=0:
            dp[r-1][c-1] = 1
        else:
            dp[r-1][c-1] = 1 - a[r-1][c-1]
        return self.helperFn(a, r-1, c-1, dp, dp[r-1][c-1])
    
    def helperFn(self, a, i, j, dp, h):
        if i == 0 and j == 0:
            dp[i][j] = h
            return dp[i][j]
        
        if i == 0:
            if dp[i][j] == 0:
                x = a[i][j-1]
                if x >= h:
                    h_x = 1
                else:
                    h_x = h - x
                dp[i][j] = self.helperFn(a, i, j-1, dp, h_x)
                
        elif j == 0:
            if dp[i][j] == 0:
                y = a[i-1] [j]
                if y >= h:
                    h_y = 1
                else:
                    h_y = h - y
                dp[i][j] = self.helperFn(a, i-1, j, dp, h_y)   
        
        else:
            x, y = a[i-1][j], a[i][j-1]
            if x >= h:
                h_x = 1
            else:
                h_x = h - x
            if y >= h:
                h_y = 1
            else:
                h_y = h - y
            dp[i][j] = min(self.helperFn(a, i-1, j, dp, h_x), self.helperFn(a, i, j-1, dp, h_y))
        return dp[i][j]
        
o = Solution()
A = [

   [1, -1, 0],

   [-1, 1, -1],

   [1, 0, -1]

 ]
print(o.calculateMinimumHP(A))
```

```python
# IterativeApproach --> BottomUp DP
class Solution:
    def calculateMinimumHP(self, d):
        row_len, col_len = len(d), len(d[0])
        DP = None
        for i in range(row_len-1, -1, -1):
            newDP = [None]*(col_len)
            for j in range(col_len-1, -1, -1):
                if i == row_len-1 and j == col_len-1:
                    tmp = 1
                else:
                    tmp = float('inf')
                if j < col_len - 1:
                    tmp = min(newDP[j+1], tmp)
                if i < row_len - 1:
                    tmp = min(DP[j], tmp)
                newDP[j] = max(tmp - d[i][j], 1)
            DP = newDP
        return DP[0]
```
