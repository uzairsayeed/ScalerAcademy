## Add Two Numbers as Lists:
Problem Description

You are given two linked lists, A and B representing two non-negative numbers.

The digits are stored in reverse order and each of their nodes contain a single digit.

Add the two numbers and return it as a linked list.
### Problem Constraints
1 <= |A|, |B| <= 105
### Input Format
The first argument of input contains a pointer to the head of linked list A.

The second argument of input contains a pointer to the head of linked list B.
### Output Format
Return a pointer to the head of the required linked list.
## Example Input
### Input 1:
 A = [2, 4, 3]
 
 B = [5, 6, 4]
### Input 2:
 A = [9, 9]
 
 B = [1]
## Example Output
### Output 1:
 [7, 0, 8]
### Output 2:
 [0, 0, 1]
## Example Explanation
### Explanation 1:
 A = 342 and B = 465. A + B = 807. 
### Explanation 2:
 A = 99 and B = 1. A + B = 100. 

```python
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None

def printll(head):
    list = []
    while head:
        list.append(head.val)
        head = head.next
    print(list)
    #return list    

def ll(arr):
    if len(arr)==0:
        return None
    head = ListNode(arr[0])
    last = head
    for val in arr[1:]:
        last.next = ListNode(val)
        last = last.next
    return head

class Solution:
    # @param A : head node of linked list
    # @param B : head node of linked list
    # @return the head node in the linked list
    def addTwoNumbers(self, head1, head2):
        p1, p2 = head1, head2
        dummy = ListNode(None)
        p = dummy
        carry = 0
        while p1 and p2:
            curr = p1.val + p2.val + carry
            #print(curr)
            carry = 0
            if curr >= 10:
                t = curr
                #print(curr, carry)
                curr %= 10
                carry = t // 10
            curr_node = ListNode(curr)
            p.next = curr_node
            p = p.next
            #printll(p)
            p1 = p1.next
            p2 = p2.next
        #printll(p)
        #print(carry)
        #printll(p1)
        while p1:
            curr = p1.val + carry
            carry = 0
            if curr >= 10:
                t = curr
                curr %= 10
                carry = t // 10
            curr_node = ListNode(curr)
            p.next = curr_node
            p = p.next
            p1 = p1.next
        while p2:
            curr = p2.val + carry
            carry = 0
            if curr >= 10:
                t = curr
                curr %= 10
                carry = t//10
            curr_node = ListNode(curr)
            p.next = curr_node
            p = p.next
            p2 = p2.next
        if carry:
            p.next = ListNode(carry)
            p = p.next
            carry = 0
        return dummy.next
            
# To accept input for LL    
arr1=list(int(i) for i in input().strip().split(' '))
arr2=list(int(i) for i in input().strip().split(' '))
# Passing the input list elements to create LL 
# Note: End your input with -1
# If you want to create 1->2->3->4->5->None as your LL then pass 1 2 3 4 5 -1 as input
a = ll(arr1[:-1])
b = ll(arr2[:-1])
# Creating object of Solution class
o = Solution()
# Calling the inside function of Solution class with 'o' created object
x = o.addTwoNumbers(a, b)
printll(x)
```
