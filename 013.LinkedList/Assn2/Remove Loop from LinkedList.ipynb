## Remove Loop from Linked List:
Problem Description

Given a linked list which contains some loop.

You need to find the node, which creates a loop, and break it by making the node point to NULL.
### Problem Constraints
1 <= number of nodes <= 1000
### Input Format
Only argument is the head of the linked list.
### Output Format
return the head of the updated linked list.
## Example Input
### Input 1:
 
        1 -> 2

        ^    |

        | - - 

### Input 2:

3 -> 2 -> 4 -> 5 -> 6
          
          ^         |
          
          |         |    
          
          - - - - - -
## Example Output
### Output 1:
 1 -> 2 -> NULL
### Output 2:
 3 -> 2 -> 4 -> 5 -> 6 -> NULL
## Example Explanation
### Explanation 1:
 Chain of 1->2 is broken.
### Explanation 2:
 Chain of 4->6 is broken.

```python
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None
def printll(head):
    print(123)
    list = []
    while head:
        list.append(head.val)
        head = head.next
    print(list)
    #return list    
def ll(arr):
    if len(arr)==0:
        return None
    head = ListNode(arr[0])
    last = head
    for val in arr[1:]:
        last.next = ListNode(val)
        last = last.next
    return head
class Solution:
    # @param A : head node of linked list
    # @return the head node in the linked list
    def solve(self, head):
        if not head or not head.next:
            return head
        slow, fast = head, head
        bFlag = False
        # Detecting whether loop exists or not
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            # If loop exists at some pt. slow == fast
            if slow == fast:
                bFlag = True
                break
        # If there's no cycle return original head
        if not bFlag:
            return head
        # Now find the length of cycle
        n = 1
        slow = slow.next
        while slow != fast:
            slow = slow.next
            n+=1
        # Now take 2 ptrs st, end and place them 'n' distance apart starting from head
        # prev pointer points to the node prev to end pointer
        st, end, prev = head, head, None
        while n!=0:
            prev = end
            end = end.next
            n-=1
        while st != end:
            prev = end
            st = st.next
            end = end.next
        prev.next = None
        return head
arr=list(int(i) for i in input().strip().split(' '))
l = ll(arr[:-1])
o = Solution()
x = o.sortList(l)
printll(x)

#8 3 4 7 1 0 5 -1
```
