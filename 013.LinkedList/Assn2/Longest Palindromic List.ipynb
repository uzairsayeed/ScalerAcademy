## Longest Palindromic List:
Problem Description

Given a linked list of integers. Find and return the length of the longest palindrome list that exists in that linked list.

A palindrome list is a list that reads the same backward and forward.

Expected memory complexity : O(1)
### Problem Constraints
1 <= length of the linked list <= 2000

1 <= Node value <= 100
### Input Format
The only argument given is head pointer of the linked list.
### Output Format
Return the length of the longest palindrome list.
## Example Input
### Input 1:
 2 -> 3 -> 3 -> 3
### Input 2:
 2 -> 1 -> 2 -> 1 ->  2 -> 2 -> 1 -> 3 -> 2 -> 2
## Example Output
### Output 1:
 3
### Output 2:
 5
## Example Explanation
### Explanation 1:
 3 -> 3 -> 3 is largest palindromic sublist
### Explanation 2:
 2 -> 1 -> 2 -> 1 -> 2 is largest palindromic sublist.


```python
# Definition for singly-linked list.
# class ListNode:
#    def __init__(self, x):
#        self.val = x
#        self.next = None

class Solution:
    # @param A : head node of linked list
    # @return an integer
    def solve(self, head):
        if not head.next:
            return 1
        ans = 1
        # Calculating for odd length palindromes
        prev, curr = None, head
        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
            # Now as we've set our centre which is the node @ prev
            # Start comparing in left n right directions respectively
            # Odd lenght paindromes will have only 1 center
            # .....x2 <- x1 <- [c1]  y1 -> y2..... 
            p1, p2 = prev.next, curr
            curr_odd_len = 1
            while p1 and p2:
                if p1.val != p2.val:
                    break
                curr_odd_len += 2
                p1 = p1.next
                p2 = p2.next
            ans = max(curr_odd_len, ans)
        # Now as we've processed all the odd length palindromes
        # Reverse the existing LL
        # Continue the same above process for even length palindromes
        head = prev
        prev, curr = None, head
        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
        # Now LL is back to its original state
        # Setting head to prev
        head = prev
        prev, curr = None, head
        # Calculating for even length palindromes
        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
            # Even length palindromes will have 2 centers 
            # .....x2 <- x1 <- [c1]  [c2] -> y1 -> y2.....i
            if prev and curr and prev.val == curr.val:
                p1, p2 = prev.next, curr.next
                curr_even_len = 2
                while p1 and p2:
                    if p1.val != p2.val:
                        break
                    curr_even_len += 2
                    p1 = p1.next
                    p2 = p2.next
                ans = max(curr_even_len, ans)
        return ans              
```
