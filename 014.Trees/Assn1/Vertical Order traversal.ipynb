## Vertical Order traversal:
Problem Description

Given a binary tree, return a 2-D array with vertical order traversal of it. Go through the example and image for more details.
### NOTE:
If 2 Tree Nodes shares the same vertical level then the one with lesser depth will come first.
### Problem Constraints
1 <= number of nodes <= 105
### Input Format
First and only arument is a pointer to the root node of binary tree, A.
### Output Format
Return a 2D array denoting the vertical order traversal of tree as shown.
## Example Input
### Input 1:

          6
        
        /   \
       
       3     7
      
      / \     \
     
     2   5     9
### Input 2:

          1
        
        /   \
       
       3     7
      
      /       \
     
     2         9
## Example Output
### Output 1:

 [
    
    [2],
    
    [3],
    
    [6, 5],
    
    [7],
    
    [9]
 
 ]
### Output 2:

 [
    
    [2],
    
    [3],
    
    [1],
    
    [7],
    
    [9]
 
 ]
## Example Explanation
### Explanation 1:
 First row represent the verical line 1 and so on.

```python
# Definition for a  binary tree node
# class TreeNode:
#	def __init__(self, x):
#		self.val = x
#		self.left = None
#		self.right = None

class Solution:
    # @param A : root node of tree
    # @return a list of list of integers
    def verticalOrderTraversal(self, root):
        if not root:
            return []
        map = {}
        
        def traverse(root, d, l):
            if not root:
                return 
            if d in map:
                map[d].append([l, root.val])
            else:
                map[d] = [[l, root.val]]
            traverse(root.left, d-1, l+1)
            traverse(root.right, d+1, l+1)
            
        traverse(root, 0, 0)
        _min, _max = min(map.keys()), max(map.keys())
        res = []
        for i in range(_min, _max+1):
            curr = sorted(map[i], key = lambda x: x[0])
            res.append(list(node for level, node in curr))
        return res
```
