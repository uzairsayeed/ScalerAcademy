## Next Pointer Binary Tree:
Problem Description

Given a binary tree,

Populate each next pointer to point to its next right node. If there is no next right node, the next pointer should be set to NULL.

Initially, all next pointers are set to NULL.

Assume perfect binary tree and try to solve this in constant extra space.
### Problem Constraints
1 <= Number of nodes in binary tree <= 100000

0 <= node values <= 10^9
### Input Format
First and only argument is head of the binary tree A.
### Output Format
Return the head of the binary tree after the changes are made.
## Example Input
### Input 1:

 
     1
    
    /  \
   
   2    3
### Input 2:
 
        1
       
       /  \
      
      2    5
     
     / \  / \
    
    3  4  6  7
## Example Output
### Output 1:
 
        1 -> NULL
       
       /  \
      
      2 -> 3 -> NULL
### Output 2:
 
         1 -> NULL
       
       /  \
      
      2 -> 5 -> NULL
     
     / \  / \
    
    3->4->6->7 -> NULL
## Example Explanation
### Explanation 1:
Next pointers are set as given in the output.
### Explanation 2:
Next pointers are set as given in the output.


```python
# Definition for a  binary tree node
# class TreeLinkNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
#         self.next = None

class Solution:
    # @param root, a tree node
    # @return nothing
    def connect(self, root):
        if not root:
            return None
        root.next = None
        s = [root]
        while s:
            curr = s.pop()
            if curr.left:
                if curr.right:
                    curr.left.next = curr.right
                else:
                    t = curr.next
                    while t:    
                        if t.left:
                            curr.left.next = t.left
                            break
                        elif curr.next.right:
                            curr.left.next = t.right 
                            break
                        else:
                            t = t.next
                    if not t:
                        curr.left.next = None
            if curr.right:
                if curr.next:
                    if curr.next.left:
                        curr.right.next = curr.next.left
                    else:
                        t = curr.next
                        while t:
                            if t.left:
                                curr.right.next = t.left
                                break
                            elif t.right:
                                curr.right.next = t.right
                                break
                            else:
                                t = t.next
                        if not t:
                            curr.right.next = None
                else:
                    curr.right.next = None
            if curr.right:
                s.append(curr.right)
            if curr.left: 
                s.append(curr.left)
        return root
```

```python
class Solution:
    # @param root, a tree node
    # @return nothing
    def connect(self, root):
        if not root:
            return root
        
        pre = root
        
        while pre:
            curr = pre
            while curr:
                if curr.left:
                    curr.left.next = curr.right
                    if curr.next:
                        curr.right.next = curr.next.left
                curr = curr.next
            pre = pre.left
        
        return root
```
