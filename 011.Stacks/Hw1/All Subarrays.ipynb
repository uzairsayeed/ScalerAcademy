## All Subarrays:
Problem Description

Given an integer array A of size N. You have to generate it's all subarrays having the size greater than 1.

Then for each subarray find Bitwise XOR of its maximum and second maximum element.

Find and return the maximum value of XOR among all subarrays.
### Problem Constraints
2 <= N <= 105

1 <= A[i] <= 107
### Input Format
Only argument is an integer array A.
### Output Format
Return an integer, i.e maximum value of XOR of maximum and 2nd maximum element among all subarrays.
## Example Input
### Input 1:
 A = [2, 3, 1, 4]
### Input 2:
 A = [1, 3]
## Example Output
### Output 1:
 7
### Output 2:
 2
## Example Explanation
### Explanation 1:
 All subarrays of A having size greater than 1 are:
 
 Subarray            XOR of maximum and 2nd maximum no.
 
 1. [2, 3]           1
 
 2. [2, 3, 1]        1
 
 3. [2, 3, 1, 4]     7
 
 4. [3, 1]           2
 
 5. [3, 1, 4]        7
 
 6. [1, 4]           5

So maximum value of Xor among all subarrays is 7.
### Explanation 2:
 Only subarray is [1, 3] and XOR of maximum and 2nd maximum is 2.

```python
class Solution:
    # @param A : list of integers
    # @return an integer
    def solve(self, a):
        n = len(a)     
        ans = -1
        l = self.next_largest_from_left(a)
        r = self.next_largest_from_right(a)
        #print(l, r)
        for i in range(n):
            curr = 0
            if i == 0:
                if r[i] < n:
                    curr = a[i] ^ a[r[i]]
            elif i == n-1:
                if l[i] >= 0:
                    curr = a[i] ^ a[l[i]]
            else:
                curr1, curr2 = 0, 0
                if r[i] < n:
                    curr1 = a[i] ^ a[r[i]]
                if l[i] >= 0:
                    curr2 = a[i] ^ a[l[i]]
                curr = max(curr1, curr2)
            ans = max (ans, curr)
        return ans
    def next_largest_from_left(self, a):
        n = len(a)
        res = [-1]*n
        s = []
        s.append(0)
        for i in range(1, n):
            if a[i] < a[s[-1]]:
                res[i] = s[-1]
            else:
                while len(s) != 0 and a[s[-1]] < a[i]:
                    s.pop()
                if len(s) != 0:
                    res[i] = s[-1]
            s.append(i)
        return res
    def next_largest_from_right(self, a):
        n = len(a)
        s = []
        res = [n]*n
        s.append(0)
        for i in range(1, n):
            if a[i] < a[s[-1]]:
                s.append(i)
            else:
                while len(s)!=0 and a[i] > a[s[-1]]:
                    res[s.pop()] = i
                s.append(i)
        return res
o = Solution()
a = [1,1]
print(o.solve(a))
```
