```python
class Solution:
    # @param A : list of integers
    # @param B : integer
    # @return an integer
    def solve(self, a, b):
        n = len(a)
        res = []
        for i in range(n-2):
            for j in range(i+1, n-1):
                for k in range(j+1, n):
                    res.append(a[i]+a[j]+a[k])
        res.sort()
        print(res)
        return res[b-1]

o = Solution()
a = [22,10,5,11,16,2,11,7,16,2,17,6,15,3,9,6]
b = [18,23,11,26]
print(o.solve(a, 183))
        
```

```python
import heapq
class Solution:
    # @param A : list of integers
    # @param B : integer
    # @return an integer
    def solve(self, a, b):
        n = len(a)
        res = []
        h = []
        psum = []
        for i in range(n-1):
            for j in range(n):
                psum.append([a[i], a[i]+a[j]])
        #print(psum)
        for i in range(n-1):
            for j in range(n):
                if a[i] != psum[j][0]:
                    curr = -1 * (a[i]+psum[j][1])
                    print(h, curr)
                    if len(h) <= b:
                        heapq.heappush(h, -1 * curr)
                    else:
                        curr *= -1
                        if curr >= h[0]:
                            heapq.heapreplace(h, curr)
        heapq.heappop(h)
        print(h, h[0])
o = Solution()
a = [18,23,11,16,5,23,7,20,20,10]
ba = [22,10,5,11,16,2,11,7,16,2,17,6,15,3,9,6]
print(o.solve(a, 105))
```
